```python
import numpy as np
import math as math

from tools._graphtools import *
from tools._vectools import * 
from matplotlib import colors

np.set_printoptions(linewidth=np.inf,precision=3, suppress=True)
```

```python
data = np.load(f"walkresults/2avg_rc_2.npz",allow_pickle=True)
t = data["t"]
prob_bare = data["prob_bare"]
prob_flux_dict = data["prob_flux_dict"].item()

flux = list(prob_flux_dict.keys())

prob_mat = list(prob_flux_dict.values())

heatmap_data = [i-prob_bare for i in prob_mat]

heatmap_array = np.array(heatmap_data)

# Create the figure and axes
fig, ax = plt.subplots()

# Plot using imshow, with no interpolation
# 'origin="lower"' makes the first row of data correspond to flux[0] at the bottom
im = ax.imshow(heatmap_array, 
               interpolation='none', 
               aspect='auto', 
               origin='lower',
               extent=[0, t.max(), 0, 2])

# Optional: if 'flux' is numeric and sorted ascending, you can use 'extent' to set continuous x/y-limits
# e.g., extent=[t[0], t[-1], flux[0], flux[-1]]
# Otherwise, for discrete ticks, set them manually:

# Label x-axis with time, y-axis with flux
# (If t or flux is very large, you might want fewer ticks or other spacing strategies)
ax.set_xticks( list(range(0, int(t.max()))))
ax.set_yticks(np.linspace(0,1,11))

ax.set_xlabel("Time")
ax.set_ylabel(r"$\Phi/2\pi$",fontsize=20)

# Add a labeled color bar
cbar = fig.colorbar(im, ax=ax)
cbar.set_label(r"$|\langle N | \Psi \rangle |_{\Phi=0} - |\langle N | \Psi \rangle |_{\Phi=\phi}$",fontsize=20)
plt.title("Difference Between Fluxed and Unfluxed (d=2)")

plt.show()
```

```python
data = np.load(f"walkresults/2043311644_30000_MC.npz",allow_pickle=True)
t = data["t"]
prob_bare = data["prob_bare"]
prob_flux_dict = data["prob_flux_dict"].item()

flux = list(prob_flux_dict.keys())

prob_mat = list(prob_flux_dict.values())
heatmap_array = np.array(prob_mat)

# Create the figure and axes
fig, ax = plt.subplots()

log_norm = colors.SymLogNorm(linthresh=1e-6, linscale=1.0,vmin=heatmap_array.min(), vmax=heatmap_array.max())

im = ax.imshow(heatmap_array,
               interpolation='none',
               aspect='auto',
               origin='lower',
               extent=[0, t.max(), 0, 2],
               norm=log_norm)      # pass the LogNorm here

ax.set_xticks( list(range(0, int(t.max()), int(t.max()/10) )))
ax.set_yticks(np.linspace(0,2,11))

ax.set_xlabel("Time")
ax.set_ylabel(r"$\Phi/2\pi$",fontsize=20)

# Add a labeled color bar
cbar = fig.colorbar(im, ax=ax)
cbar.set_label(r"$|\langle N | \Psi \rangle |_{\Phi=\phi}$",fontsize=20)
plt.title("Binary Tree Averaged CRWs (d=3)")
plt.show()
```
