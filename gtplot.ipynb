```python
import numpy as np

from graphtools import *
from vectools import *
```

```python
max_depth = 3
p=4

vertex_sizes = [600,400,50,20]
o_w = [4,3.5,1.6]
e_w = [7,5,3,1.2]
x_len = [3,6,9]


#fig, axes = plt.subplots(1, max_depth, figsize=(20, 6))  # adjust figsize as needed

fig, axes = plt.subplots(nrows=1, ncols=max_depth, figsize=(sum(x_len), 6),
                         gridspec_kw={'width_ratios': x_len})


# Loop over each subplot in the grid:
for j in range(1,max_depth+1):
    # For demonstration, create tree positions and a tree graph for each subplot
    X = [p]*j
    positions = tree_coords(X, xd=1, yd=1)
    tree = generate_good_tree(X)  # Replace with your actual tree generation
    tree_copy = tree.deep_copy()
    tree_size =  len(tree_copy.nodes)
    bot_layer = [int(node.index) for node in tree.nodes if node.degree == 1]
    copy_bot_layer = [int(leaf) + tree_size for leaf in bot_layer]

    gt_illusion = ConnectedGraph(nodes = tree_copy.nodes | shift_graph(tree, tree_size ).nodes )
    
    # Plot the tree on the corresponding Axes
    ax = pl_graph(axes[j-1], gt_illusion, positions= glued_tree_coords(X,xd=x_len[j-1] ,yd=4,spacing=0), 
                    vertex_size=vertex_sizes[j-1],outline_weight = o_w[j-1], edge_weight=e_w[j-1],margin=.5)


# Adjust layout to prevent overlap and show the combined figure:
plt.tight_layout()
plt.show()
```
