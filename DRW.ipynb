```python
import numpy as np

from graphtools import *
from vectools import * 

from numpy.linalg import matrix_power
```

```python
X = [2]*5
AFB = 2*np.pi/X[0]

ftree = fluxedTree(X)
ftree.construct_fluxed()

rand = generate_random_cycle_graph(ftree)
rand.construct_fluxed()
rand.construct_adj()
```

```python
bare_hamiltonian = rand.adj
fluxed_hamiltonian = rand.weighted_adj(AFB)
```

```python
N = len(rand.node_map)
psi_i = e_n(0, N)

end = e_n(N-1,N)
```

```python
t = list(range(0, 20))

prob_flux = []
prob_bare = []
for time in t:
    prob_bare.append(abs(end@matrix_power(bare_hamiltonian,time)@psi_i)**2 )
    prob_flux.append(abs(end@matrix_power(fluxed_hamiltonian,time)@psi_i)**2 )
```

```python
plt.plot(t, prob_bare,label=r"$\Phi = 0$")
plt.plot(t, prob_flux,label=r"$\Phi = \pi$")

plt.legend(fontsize=15)
```
