```python
X =[2]
```

```python
ftree = generate_ft(X)
rand = generate_rgc(ftree)
rand.construct_adj()
```

```python
fig, axes = plt.subplots(1, 1, figsize=(8, 8))  # adjust figsize as needed

pl_graph(axes, rand, positions=rand_cycle_coords(X,xd=3,yd=1,spacing=1), vertex_size=60,outline_weight = 1, edge_weight=2,margin=.2)

plt.show()
```

```python
from tools._walktools import tbw

kvalues = np.linspace(-np.pi, np.pi,10)
rand.construct_period(0)

bands = [] 
for k in kvalues:
    bands.append(np.linalg.eigvalsh(rand.period(k)))

bands = np.transpose(bands)

for band in bands:
    plt.plot(kvalues,band,color="black")

tbw(bands)
```

```python
data = np.load(f"bw/2741489904_20000_MCbw.npz",allow_pickle=True)
fluxes = data["fluxes"]
bw = data["avgbw"]
neck = data["necklaces"]
X = data['sequence']

fs = 20
plt.plot(fluxes,bw,label="N = %d" %len(neck.item()))


plt.xticks(
        np.linspace(0, 4*np.pi, 5),
        np.linspace(0,       2, 5))
#for p in FBP:
#        plt.axvline(p)
plt.xlabel(r"$\Phi/2 \pi$",fontsize=fs)
plt.ylabel("Bandwidth (J)",fontsize=fs)
plt.title(X)
plt.legend()
```
